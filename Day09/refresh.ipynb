‚ö†Ô∏è Rules:

Solve it with a for loop first.

Then rewrite using a comprehension + sum().


```python
text = "programming is power"
new = ""
for char in text:
     if char in ['a', 'e','i', 'o', 'u']:
          new += char

print(len(new))

#list comprehension
new_list = [char for char in text if char in ['a', 'e', 'i', 'o', 'u']]
print(new_list)
print(len(new_list))

#optimal
count = sum(1 for char in text if char in "aeiou") #for every vowel found, yield the number 1 and sum them all up
print(count)
```

Count Palindromes
words = ["racecar", "madam", "apple", "noon", "python"]


üëâ Count how many words are palindromes.

```python
words = ["racecar", "madam", "apple", "noon", "python"]

word_count = sum(1 for word in words if word == word[::-1] )
print(word_count)

# alternative, less efficient, less safe 
word_count2 = sum(words.count(word) for word in words if word == word[::-1])
print(word_count2)

#readable

word_list = []
for word in words:
    if word == word[::-1]:
        word_list.append(word)

print(word_list)
print(len(word_list))
```

### **Rules for Prime Numbers**

1. **Definition**:
   A prime number is a number greater than 1 that has **exactly two divisors**:

   * **1**
   * **itself**

   Example: `5` ‚Üí divisors are `{1, 5}` ‚Üí ‚úÖ prime.

---

2. **Special Cases**:

   * `0` ‚Üí **not prime**
   * `1` ‚Üí **not prime** (it has only one divisor, not two)
   * `2` ‚Üí **prime** (smallest prime number, the only even prime)

---

3. **Divisibility Test**:
   A number `n` is **not prime** if it can be divided evenly by any number other than `1` or `n`.

   * Example: `9` ‚Üí divisible by `3` ‚Üí ‚ùå not prime.
   * Example: `13` ‚Üí only divisible by `1` and `13` ‚Üí ‚úÖ prime.

---

4. **Optimization Rules** (useful in coding):

   * **Skip checking even numbers** (except `2`) ‚Üí they‚Äôll never be prime.
   * **Check only up to ‚àön**:
     If a number `n` has a factor larger than ‚àön, its matching pair is smaller than ‚àön.
     ‚Üí No need to check all the way up to `n-1`.

   Example:

   * To test if `29` is prime ‚Üí ‚àö29 ‚âà 5.38 ‚Üí check divisors only up to `5`.

---

‚ö° So the **prime rules condensed**:

* Prime > 1.
* Divisors = exactly 2 (1 and itself).
* `2` is the only even prime.
* Efficient check: test divisibility up to ‚àön.

---

```python
#checker
import math
def is_prime(n):
    if n < 2: # 0 and 1 are not prime
        return False
    for i in range(2, int(n**0.5) + 1): # start with 2 (only even prime), n**0.5 for taking square root, + 1 to inlcude the integer just past the square root
        if n % i == 0: #if any i divides n perfectly, then n is not prime
            return False
    return True

nums = range(1,51)
prime_count = sum(1 for n in nums if is_prime(n))
print(prime_count)
prime_list = [n for n in nums if is_prime(n)]
print(prime_list)
```

Prime Range Finder

üîπ Goal: Write a function that takes two numbers a and b, and returns all prime numbers between them.

Hint: Use your is_prime function inside a loop over range(a, b+1).

Extra: Try it with a = 10, b = 50.

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def prime_range(a, b):
    primes = []
    for num in range(a, b + 1): #gives loop  given interval a,b
        if is_prime(num):
            primes.append(num)
    return primes

print(prime_range(10, 50))
```
