```python
# Write a function that calculates tip
# tip_calculator(bill_amount, tip_percentage)
# Example: tip_calculator(50, 18) should return 9.0


def tip(bill_amount, tip_percentage):
    tip = bill_amount * (tip_percentage/100)
    return tip

tip(50, 18)
```

```python
# Write a function that takes a name and makes it "fancy"
# fancy_name("john") should return "✨ JOHN ✨"

def fancy_name(name):
    return f"✨{name.upper()}✨"
print(fancy_name("john"))
```

```python
# Given a list of numbers, return only the even ones
# filter_evens([1, 2, 3, 4, 5, 6]) should return [2, 4, 6]


def even_numbers(numbers):
    even = []
    for i in numbers:
        if i % 2 == 0:
            even.append(i)
    return even

print(even_numbers([1, 2, 3, 4, 5, 6]))
```

```python
# Create a simple grade book
# Write a function to find the highest grade

grades = {"alice": 85, "bob": 92, "charlie": 78}

def highest_grade(grades):
    highest = -1
    for grade in grades.values():
        if grade > highest:
            highest = grade
    return highest

print(highest_grade(grades))

```

```python
# Make a simple calculator that asks for two numbers
# and an operation (+, -, *, /) then shows the result

def calculate(number1, number2, operation):
    if operation == '+':
        add = number1 + number2
        return add
    elif operation == '-':
        subtract = number1 - number2
        return subtract
    elif operation == '*':
        multiply = number1 * number2
        return multiply
    elif operation == '/':
        divide = number1 / number2
        return divide
    else:
        return "Invalid Operation, try again"
    
# asking for input

def ask():
    number1 = int(input("Enter first number: "))
    number2 = int(input("Enter second number: "))
    operation = input("Enter operation (+, -, *, /): ")

    return number1, number2, operation

number1, number2, operation = ask()

print(calculate(number1, number2, operation))
```

```python
# Write a function that counts down from any number
# countdown(5) should print: 5, 4, 3, 2, 1, "Blast off!"

def countdown(number):
    count = []
    for i in range(number, 0, -1):
        count.append(i)
    
    for num in count:
        print(num, end = ", ")
    print('"Blast off!"')


countdown(5)
```

```python
# Check if password is "strong":
# - At least 8 characters
# - Contains a number
# - Contains uppercase and lowercase

def check_password(password):
    if len(password) < 8:
        return False
    
    has_number = False
    has_upper = False
    has_lower = False

    for char in password:
        if char.isdigit():
            has_number = True
        if char.isupper():
            has_upper = True
        if char.islower():
            has_lower = True

    return has_number and has_upper and has_lower
        
print(check_password("Password1"))    # Should be True
print(check_password("password"))     # Should be False (no number, no uppercase)
print(check_password("PASSWORD1"))    # Should be False (no lowercase)
print(check_password("Pass1"))        # Should be False (too short)
```
