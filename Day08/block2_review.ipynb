## üß© **Challenge: Clean the Usernames!**

### üìú Scenario:

You're building a user registration system. You receive a raw list of submitted usernames, but not all of them are valid.

### ‚úÖ **Rules for Valid Usernames:**

1. Must be **at least 5 characters long**.
2. Must **not contain any digits**.
3. Must be **all lowercase**.

---

### üß† Task:

Write a program that takes this list:

```python
usernames = [
    "admin", "Guest123", "john", "alice", "eve99", "root42",
    "Zoey", "charlie", "b0bby", "delta", "gamma1", "lambda"
]
```

And produces a new list:

```python
valid_usernames = [...]  # only those that follow all 3 rules
```

### üß™ Expected Output:

```python
['admin', 'alice', 'charlie', 'delta', 'lambda']
```

---

## ‚úçÔ∏è Bonus:

* Try solving it **first using a `for` loop + append**.
* Then try rewriting it as a **list comprehension**.

---


````python
usernames = [
    "admin", "Guest123", "john", "alice", "eve99", "root42",
    "Zoey", "charlie", "b0bby", "delta", "gamma1", "lambda"
]

valid_usernames = []

# using for loop + append

for word in usernames:
    has_digit = False

    for char in word:
        if char.isdigit():
            has_digit = True
            break #stops as soon as we know the answer
    
    if not has_digit:
        valid_usernames.append(word)

print(valid_usernames)


#list comprehension

filtered = [word for word in usernames if not any(char.isdigit() for char in word)]
print(filtered)
````

```python
inventory = [
    {"item": "potion", "qty": 2},
    {"item": "scroll", "qty": 1},
    {"item": "dagger", "qty": 3}
]
```

Tasks:

- Find total quantity of all items
    
- Increase potion qty by 1
    
- Remove all daggers
    
- Print: `"You have 3 potions, 1 scroll"`
    

‚úîÔ∏è Requires: loops, conditionals, list modification

````python

inventory = [
    {"item": "potion", "qty": 2},
    {"item": "scroll", "qty": 1},
    {"item": "dagger", "qty": 3}
]

# Find total quantity of all items

for item in inventory:
    qty = sum(item['qty'] for item in inventory)

print(qty)

# Increase potion qty by 1
for item in inventory:
    item['qty'] += 1
    print(item)


# - Remove all daggers
for item in inventory:
    if item['item'] == 'dagger':
        inventory.remove(item)
print(inventory)

#list comprehension

new_inventory = [item for item in inventory if item['item'] != 'dagger']
print(new_inventory)  




````

Love Calculator
üí™ This is a difficult challenge! üí™ (From App Brewery)

You are going to write a function called calculate_love_score() that tests the compatibility between two names.  To work out the love score between two people: 

1. Take both people's names and check for the number of times the letters in the word TRUE occurs.   

2. Then check for the number of times the letters in the word LOVE occurs.   

3. Then combine these numbers to make a 2 digit number and print it out. 

e.g.

name1 = "Angela Yu" name2 = "Jack Bauer"

T occurs 0 times 

R occurs 1 time 

U occurs 2 times 

E occurs 2 times 

Total = 5 

L occurs 1 time 

O occurs 0 times 

V occurs 0 times 

E occurs 2 times 

Total = 3 

Love Score = 53

Example Input 

calculate_love_score("Kanye West", "Kim Kardashian")

Example Output

42

````python
def calculate_love_score(name1, name2):
    combined = (name1 + name2).lower()
    total_true = sum(combined.count(char) for char in 'true')
    total_love = sum(combined.count(char) for char in 'love')
    

    print(f"{total_true}{total_love}")
    

calculate_love_score("Kanye West", "Kim Kardashian")
````
